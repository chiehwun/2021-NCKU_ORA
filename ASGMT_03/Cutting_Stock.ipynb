## Compare m
### Reduced Problem 3

```python
R = Rects()
R.add(8, 16, 1)
R.add(9, 9, 1)
R.add(18, 3, 3)
```

```python
m = 25
t = f'Reduced Problem 3 (m={m})'
%timeit Cutting_Stock(R, m=m, same=True,graph_out=False, title=t, show_detail=False)
Cutting_Stock(R, m=m, same=True,graph_out=True, title=t, show_detail=False)
```

```python
M = [25,50,75,100]
T = [73.9, 176, 318, 593]

plt.figure()
plt.plot(M, T, '.-')
plt.title('m-t plot', fontsize=20)
plt.xlabel('m')
plt.ylabel('Time (ms)')
plt.savefig('m-t_plot.png')
plt.show()
```
